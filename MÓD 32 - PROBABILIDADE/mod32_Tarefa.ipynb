# Probabilidade - Tarefa

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import seaborn as sns
import pandas as pd

import matplotlib.ticker as mtick
from scipy.stats import norm
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Definindo os parâmetros da distribuição
média = 15  # média
desvio = 0.1  # desvio padrão

# Peso limite superior
peso_limite = 15.2

# Calculando a probabilidade acumulada até 15.2 kg
prob_acumulada = stats.norm.cdf(peso_limite, loc=média, scale=desvio)

# Calculando a probabilidade de um pacote pesar mais que 15.2 kg
prob_mais_que_15_1 = 1 - prob_acumulada

print(f"A probabilidade de um pacote pesar mais que 15,2 kg é: {prob_mais_que_15_1*100:.2f}%".replace("." , ","))
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
percentil_95 = stats.norm.ppf(0.95, loc=média, scale=desvio)

print(f"O percentil de ordem 95% para o peso dos pacotes é: {percentil_95:.2f}kg".replace("." , ","))
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui

# Definindo os parâmetros da distribuição
média = 0  # média
variancia = 1  # variância
desvio = math.sqrt(variancia) # desvio padrão

# Calculando as probabilidades para a distribuição normal padrão
prob_Y1 = stats.norm.cdf(Y, loc=média, scale=desvio)

# Calculando as probabilidades para a distribuição t-Student com 5 graus de liberdade
prob_Y2 = stats.t.cdf(Y, df=5)

# Criando um DataFrame para exibir os resultados
df = pd.DataFrame({'Y': Y,
                  'Probabilidade_Y1': [f'{x * 100:.2f}%' for x in prob_Y1],
                  'Probabilidade_Y2': [f'{x * 100:.2f}%' for x in prob_Y2],
                  })

pd.set_option('display.max_rows', 200)
df
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
#estilo do gráfico
sns.set(style="whitegrid")

#plotando figura
fig, ax = plt.subplots(figsize=(20, 9))

#plotagem dos dados
ax.plot(Y, prob_Y1, label='Distribuição Normal Padrão')
ax.plot(Y, prob_Y2, label='Distribuição t-Student com 5 graus de liberdade')

#configuração dos marcadores X e Y
ax.xaxis.set_major_locator(plt.MaxNLocator(14))
ax.yaxis.set_major_locator(plt.MaxNLocator(11))

#formatando o eixo Y em formato percenttual
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))

#adicionando rótuulos
ax.set_xlabel('Valores de Y')
ax.set_ylabel('Probabilidade')
ax.set_title('Distribuições Normal Padrão e t-Student com 5 graus de liberdade')
ax.legend()

plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y2=np.linspace(0, 1, 20)
# Seu código começa aqui

# Definindo os parâmetros da distribuição
média = 0  # média
variancia = 1  # variância
desvio = math.sqrt(variancia) # desvio padrão

# Calculando as probabilidades para a distribuição normal padrão
prob_Y1 = stats.norm.ppf(Y2, loc=média, scale=desvio)

# Calculando as probabilidades para a distribuição t-Student com 5 graus de liberdade
prob_Y2 = stats.t.ppf(Y2, df=5)

# Criando um DataFrame para exibir os resultados
df = pd.DataFrame({'Y': Y2,
                  'Probabilidade_Y1': [f'{x:.2f}%' for x in prob_Y1],
                  'Probabilidade_Y2': [f'{x:.2f}%' for x in prob_Y2],
                  })

pd.set_option('display.max_rows', 200)
df
```

6. Faça um gráfico com os dados do item anterior.

```python
#estilo do gráfico
sns.set(style="whitegrid")

#plotando figura
fig, ax = plt.subplots(figsize=(20, 9))

#plotagem dos dados
ax.plot(Y2, prob_Y1, label='Distribuição Normal Padrão')
ax.plot(Y2, prob_Y2, label='Distribuição t-Student com 5 graus de liberdade')

#configuração dos marcadores X e Y
ax.xaxis.set_major_locator(plt.MaxNLocator(20))
ax.yaxis.set_major_locator(plt.MaxNLocator(22))

#adicionando rótuulos
ax.set_xlabel('Percentis')
ax.set_ylabel('Valores Y1 e Y2')
ax.set_title('Distribuições Normal Padrão e t-Student com 5 graus de liberdade')
ax.legend()

plt.show()
```
