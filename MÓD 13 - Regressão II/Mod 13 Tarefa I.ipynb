```python
y, X = patsy.dmatrices('np.log(renda) ~ sexo + posse_de_veiculo + C(posse_de_imovel, Treatment(1)) + C(qtd_filhos) + C(tipo_renda) + C(educacao, Treatment(2)) + C(estado_civil) + C(tipo_residencia, Treatment(1)) + C(idade, Treatment(18)) + C(qt_pessoas_residencia, Treatment(2))',
               data = df)
```

```python
# REMOVENDO A MENOS SIGNIFICANTE, QUE NA MINHA OPINIÃO É
# A QUANTIDADE DE FILHOS(qtd_filhos)

y, X = patsy.dmatrices('np.log(renda) ~ C(sexo) + C(posse_de_veiculo) + C(posse_de_imovel, Treatment(1)) + C(tipo_renda) + C(educacao, Treatment(2)) + C(estado_civil) + C(tipo_residencia, Treatment(1)) + C(idade, Treatment(18)) + C(qt_pessoas_residencia, Treatment(2))',
               data = df)
```

```python
modelo = sm.OLS(y,X).fit()

modelo.summary()

#SEM ALTERAÇÕES SIGNIFICANTES
```

```python
#CONTINUAR REMOVENDO AS VARIÁVEIS
# EDUCAÇÃO, ESTADO CIVIL, TIPO DE RESIDENCIA, QTD DE PESSOAS POR RESIDÊNCIA,
# IDADE e POSSE DE VEÍCULO

y, X = patsy.dmatrices('np.log(renda) ~ C(sexo) + C(posse_de_imovel, Treatment(1)) + C(tipo_renda)',
               data = df)
```
