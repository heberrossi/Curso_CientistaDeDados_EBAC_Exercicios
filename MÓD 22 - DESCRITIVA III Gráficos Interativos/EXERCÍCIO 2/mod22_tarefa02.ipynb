#### 1. Carregue os dados da tarefa anterior

````python
df = pd.read_csv(filepath_or_buffer='ativos.csv',
                 header=[0,1], 
                 index_col=0)

df
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# 1 Faça uma média móvel em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu

df_mm = df['Adj Close'].rolling(90).mean()

df_mm
````

````python
# 2 Utilizando a fução rolling(), 
# calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima

df_mm_std = df['Adj Close'].rolling(90).std()

df_mm_std
````

````python
# 3 Monte um gráfico de linhas para as médias móveis dos ativos da sua base

df_mm_graficolinha = df_mm.stack().reset_index()
df_mm_graficolinha.columns = ['data', 'ativo', 'valor']

fig = px.line(df_mm_graficolinha, x="data", y="valor", color='ativo')
fig.show()
````

````python
# 4 Monte um gráfico de linhas para os desvios padrão em janela móvel

df_mm_graficolinha_std = df_mm_std.stack().reset_index()
df_mm_graficolinha_std.columns = ['data', 'ativo', 'valor']

fig = px.line(df_mm_graficolinha_std, x="data", y="valor", color='ativo')
fig.show()
````

5 Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

os 3 gráficos apresentam grande volatividade apresentada a partir de janeiro de 2024, quando o ativo NVDA teve uma subida bem significativa no gráfico, já os dois outros ativos se apresentam mais constantes. #### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']
ativo = ['NVDA']*len(lista_indicadores)
````

````python
colunas = list(zip(lista_indicadores, ativo))

colunas
````

````python
nvda = df.loc['2024-01-01':,colunas]
nvda.columns = lista_indicadores
````

````python
nvda
````

````python
graph = {
    'x': nvda.index,
    'open': nvda.Open,
    'close': nvda.Close,
    'high': nvda.High,
    'low': nvda.Low,
    'type': 'candlestick',
    'name': 'NVDA',
    'showlegend': True
}
````

````python
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data = [graph],
    layout_title="Ações NVDIA",
    layout=layout
)
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
